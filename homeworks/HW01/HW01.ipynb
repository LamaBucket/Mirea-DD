```python
import numpy as np

py_array = [1.2, 2.4, 3.1, 0.999, 0.221]

arr1 = np.array(py_array)

print(f"1. From Python Array: {arr1}")

arr2 = np.zeros(5)

print(f"2. Zeros Array: {arr2}")

arr3 = np.ones(5)

print(f"2. Ones Array: {arr3}")

arr4 = np.arange(1, 6, 1)

print(f"3. A Range: {arr4}")

print(f"4. Shape & DataType of py_arr: {arr1.shape} {arr1.dtype}")

print(f"5. Nth item and N-Mth items: {arr1[0]}; {arr1[1:3]}")

print(f"6.1. Vector Sum of arrs: {np.add(arr1, arr3)}")
print(f"6.2. Vector Multipl of arrs: {np.multiply(arr1, arr4)}")

print(f"7. Mean of array: {arr1.mean()}")
```
